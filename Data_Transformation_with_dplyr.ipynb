# 1. Install & Load Required Libraries

```r
# Install required packages
install.packages("tidyverse")
install.packages("nycflights13")

# Load the necessary libraries
library(tidyverse)
library(nycflights13)
```

## 2. Load the Flight Data

```r
# Load the flights dataset
data("flights")

# Display the first few rows
head(flights)
```

## 3. Filtering Rows with filter()

```r
#Select all flights on January 1st:
jan1 <- filter(flights, month == 1, day == 1)
head(jan1)
```

```r
#Filter flights with arrival delay greater than 2 hours:
delayed_flights <- filter(flights, arr_delay > 120)
head(delayed_flights)
```

```r
#If you want to print both results and save the variable
(dec25 <- filter(flights, month == 12, day == 25))
```

```r
#Comparisons
#Select flights that departed in January (month == 1)
filter(flights, month == 1)
```

```r
#Select flights that did NOT depart in January
filter(flights, month != 1)
```

```r
#Select flights from March onward (month >= 3)
filter(flights, month >= 3)
```

## Logical Operator

```r
#Select flights in January that departed from JFK
filter(flights, month == 1 & origin == "JFK")
```

```r
#Select flights that departed in January OR from JFK
filter(flights, month == 1 | origin == "JFK")
```

```r
#finds all flights that departed in November or December
filter(flights, month == 11 | month == 12)
```

```r
#De Morgan’s law: !(x & y) is the same as !x | !y, and !(x |y) is the same as !x & !y
filter(flights, !(arr_delay > 120 | dep_delay > 120))
```

## 4. Arrange Rows with arrange()

```r
#Sort flights by departure delay in descending order:
arrange(flights, desc(dep_delay))
```

```r
#Sort flights by year, month, and day:
arrange(flights, year, month, day)
```

```r
#desc() to reorder by a column in descending order:
arrange(flights, desc(arr_delay))
```

## 5. Select Columns with select()

```r
#Choose specific columns:
select(flights, year, month, day, dep_time, arr_time)
```

```r
#Exclude a column:
select(flights, -tailnum)
```

```r
#Select columns based on a pattern:
select(flights, starts_with("dep"))
```

```r
# use rename() for changing column name
rename(flights, tail_num = tailnum)
```

## 6. Create New Columns with mutate()

```r
#Add computed columns:
flights_sml <- select(flights, year:day, ends_with("delay"), distance, air_time)
mutate(flights_sml,
       gain = arr_delay - dep_delay,
       speed = distance / air_time * 60)
```

## 7. Summarize Data with summarize()

```r
#Find the average departure delay:
summarize(flights, avg_dep_delay = mean(dep_delay, na.rm = TRUE))
```

```r
#Find the average delay per day:
by_day <- group_by(flights, year, month, day)
summarize(by_day, avg_delay = mean(dep_delay, na.rm = TRUE))
```

## 8. Combining Functions with %>% (Pipe)

```r
#Using group_by() and summarize() together:
flights %>%
  group_by(dest) %>%
  summarize(
    count = n(),
    avg_dist = mean(distance, na.rm = TRUE),
    avg_delay = mean(arr_delay, na.rm = TRUE)
  ) %>%
  filter(count > 20)
```

## 9. Plot Data using ggplot2

```r
flights %>%
  group_by(dest) %>%
  summarize(
    count = n(),
    avg_delay = mean(arr_delay, na.rm = TRUE)
  ) %>%
  filter(count > 20) %>%
  ggplot(aes(x = count, y = avg_delay)) +
  geom_point()
```

```r
by_dest <- group_by(flights, dest)
delay <- summarize(by_dest,
 count = n(),
 dist = mean(distance, na.rm = TRUE),
 delay = mean(arr_delay, na.rm = TRUE)
)
delay <- filter(delay, count > 20, dest != "HNL")
```

```r
ggplot(data = delay, mapping = aes(x = dist, y = delay)) +
 geom_point(aes(size = count), alpha = 1/3) +
 geom_smooth(se = FALSE)
```

```r
delays <- flights %>%
 group_by(dest) %>%
 summarize(
 count = n(),
 dist = mean(distance, na.rm = TRUE),
 delay = mean(arr_delay, na.rm = TRUE)
) %>%
filter(count > 20, dest != "HNL")
```

## Missing Values

```r

flights %>%
 group_by(year, month, day) %>%
 summarize(mean = mean(dep_delay))
```

```r
not_cancelled <- flights %>%
 filter(!is.na(dep_delay), !is.na(arr_delay))
not_cancelled %>%
 group_by(year, month, day) %>%
 summarize(mean = mean(dep_delay))
```

## Counts

```r

delays <- not_cancelled %>%
 group_by(tailnum) %>%
 summarize(
 delay = mean(arr_delay)
 )
ggplot(data = delays, mapping = aes(x = delay)) +
 geom_freqpoly(binwidth = 10)
```

```r
delays <- not_cancelled %>%
 group_by(tailnum) %>%
 summarize(
 delay = mean(arr_delay, na.rm = TRUE),
 n = n()
 )
ggplot(data = delays, mapping = aes(x = n, y = delay)) +
 geom_point(alpha = 1/10)
```

```r
delays %>%
 filter(n > 25) %>%
 ggplot(mapping = aes(x = n, y = delay)) +
 geom_point(alpha = 1/10)
```

## What proportion of flights are delayed by more than an hour?

```r

not_cancelled %>%
 group_by(year, month, day) %>%
 summarize(hour_perc = mean(arr_delay > 60))
```

## Grouping by Multiple Variables

```r

daily <- group_by(flights, year, month, day)
(per_day <- summarize(daily, flights = n()))
```

```r
# Find all groups bigger than a threshold:
popular_dests <- flights %>%
 group_by(dest) %>%
 filter(n() > 365)
popular_dests
```

```r
# Standardize to compute per group metrics:
popular_dests %>%
 filter(arr_delay > 0) %>%
 mutate(prop_delay = arr_delay / sum(arr_delay)) %>%
 select(year:day, dest, arr_delay, prop_delay)
```
